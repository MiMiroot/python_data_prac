#### 1.Handling Missing Data

```python
import pandas as pd
import numpy as np
string_data = pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
#is.null()可以获得布尔判断值，带回原series可以
string_data[string_data.isnull()]
print('###################Filtering Out Missing Data###################')
#针对series,dropna()剔除na值,也可以.notnull()
from numpy import nan as NA
data = pd.Series([1, NA, 3.5, NA, 7])
print(data.dropna())
print(data[data.notnull()])
#对dataframe
data = pd.DataFrame([[1., 6.5, 3.], [1., NA, NA],
                      [NA, NA, NA], [NA, 6.5, 3.]])
print(data,'\n')
#dropna()会剔除所有有na的行，dopna(how='all')则所有值为na的行剔除，dropna(axis=1,how='all')所有值为na的列剔除
print(data.dropna(),'\n')
print(data.dropna( how='all'),'\n')

df = pd.DataFrame(np.random.randn(7,3))
print(df,'\n')
df.iloc[:4,1] = NA; df.iloc[:2,2] = NA; print(df,'\n')
#dopna(thresh=2)剔除含2个以上NA的行
df.dropna(thresh=2)
```

```python
print('###################Filling In Missing Data###################')
print(df.fillna(0),'\n')
#用字典对每列赋特定的值
print(df.fillna({1:0.5,2:0}))
#_是什么意思？？？？？？？？？？？？？？？？？？？？？
_ = df.fillna(0, inplace=True)
print(df,'\n')
df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA; df.iloc[4:, 2] = NA
print(df, '\n')
#method='ffill'若列中存在NA，每列分别填补na之前行的数据。但列第一个值为NA则无法填补。
print(df.fillna(method='ffill'))
df1 = pd.DataFrame(np.random.randn(7,3))
print(df1,'\n')
df1.iloc[1:4,1] = NA; df1.iloc[:2,2] = NA; df1.iloc[6,1] = NA; print(df1,'\n')
print(df1.fillna(method='ffill'),'\n')
print(df.fillna(method='ffill', limit=2),'\n\n\n')
data = pd.Series([1., NA, 3.5, NA, 7])
#用均值或者中位数来替代na
print(data.fillna(data.mean()))
```

#### 2.Data Transformation

```python
print('###################Removing Duplicates###################')
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                      'k2': [1, 1, 2, 3, 3, 4, 4]})
print(data,'\n')
print(data.duplicated())
#drop_duplicates()可添加包含列名的列表，对特定列进行去重。keep='last'出现重复时保留最后一个
print(data.drop_duplicates())
data['v1'] = np.arange(7)
print(data.drop_duplicates(['k1']))
print(data.drop_duplicates(['k1','k2'], keep = 'last'))
```

```python
print('###################Transforming Data Using a Function or Mapping###################')
#把meat_to_animal添加到data中去
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon',
                              'pastrami', 'honey ham', 'nova lox'],
                     'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
print(data,'\n')
meat_to_animal = {
      'bacon': 'pig',
      'pulled pork': 'pig',
      'pastrami': 'cow',
      'corned beef': 'cow',
      'honey ham': 'pig',
      'nova lox': 'salmon'
}
#map()可以接受函数或者字典类型的含有mapping的object
#方法1
lowercased = data['food'].str.lower()
data['animal'] = lowercased.map(meat_to_animal)
#方法2
data['food'].map(lambda x: meat_to_animal[x.lower()])

```

```python
print('###################Replacing Values###################')
data = pd.Series([1., -999., 2., -999., -1000., 3.])
#replace()可以用列表方式多个替换一个，也可以多个对应替换多个
data.replace([-999, -1000], np.nan)
data.replace([-999, -1000], [np.nan, 0])
#replace()也可以用字典的方式,key替换为value
data.replace({-999:np.nan, -1000:0})

```

```python
print('###################Renaming Axis Indexes###################')
#对axis labels和某个特征一样，可以用mapping或者函数去修改
data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                     index=['Ohio', 'Colorado', 'New York'],
                     columns=['one', 'two', 'three', 'four'])
print(data,'\n')
transform = lambda x: x[:4].upper()
data.index = data.index.map(transform)
print(data,'\n')
#rename函数，修改但不影响原来的dataset
print(data.rename(index = str.title, columns = str.upper))
#rename函数，利用字典方法,key换value的值，替换新值给axis label:
print(data.rename(index={'OHIO': 'INDIANA'},columns={'three': 'peekaboo'}), '\n')
#rename，实际修改dataset
print(data.rename(index={'OHIO': 'INDIANA'}, inplace=True), '\n')
print(data)
```

```python
print('###################Discretization and Binning###################')
#离散化数据并做bin
#pd.cut()函数，第一个参数为数据，第二个参数为离散化分类的区间端点
#pd.qcut()函数，第二个参数可是int，也可以是列表。
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
a = pd.cut(ages, bins, right=False)
print(a)  #right=False表示区间左开右闭
print(a.value_counts())
#对于这些区间，也可以重新命名
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
print(pd.cut(ages, bins, labels=group_names),'\n\n')
#如果只设定区间个数，第二个参数可设定为int，用precision=2设定小数位数，当然也可用列表自己设定quantile。
data = np.random.randn(1000)
cats = pd.qcut(data, 4)
print(cats)
print(pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]))
```

```python
print('###################Detecting and Filtering Outliers###################')
data = pd.DataFrame(np.random.randn(1000,4))
print(data.describe())
#想要对abs(值)>3的值找出来并替换
data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
#筛选有至少一个超过abs(x)>3的值的行
data[(np.abs(data) > 3).any(1)]
```

```python
print('###################Permutation and Random Sampling###################')
df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
sampler = np.random.permutation(5)
print(sampler)
#根据sampler对特定行排序
df.take(sampler)
print(df.take(sampler))
#用sample函数，不replace
df.sample(n=3)
#sample函数，replace
choices = pd.Series([5, 7, -1, 6, 4])
draws = choices.sample(n=10, replace=True)
draws
```

```python
print('###################Computing Indicator/Dummy Variables###################')
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
print(df, '\n')
print(pd.get_dummies(df['key']))
dummies = pd.get_dummies(df['key'], prefix='key')
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
#整合pd.cut()和get_dummies()函数
np.random.seed(12345)
values = np.random.rand(10)
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
pd.get_dummies(pd.cut(values, bins))
```

```python
print('###################Regular Expressions###################')
#re模块3个功能：模式识别，替换，分裂
#替换
import re
text = "foo bar\t baz \tqux"
print(re.split('\s+', text))
#可以先compile好，再用。
regex = re.compile('\s+')
regex.split(text)
regex.findall(text) #输出所有match regex的pattern

#findall可以返回一个string里所有的matches，search只能返回第一个match，match仅当string从一开始匹配时才能得到结果。
text = """Dave dave@google.com
    Steve steve@gmail.com
    Rob rob@gmail.com
    Ryan ryan@yahoo.com
    """
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
regex = re.compile(pattern, flags = re.IGNORECASE)
print(regex.findall(text))
print(regex.search(text))
print(regex.match(text))
#sub可以替换对应的所有matches
print(regex.sub('REDACTED',text),'\n\n\n')

#当如果要把email的内容分段，构建pattern的时候要加括号
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern, flags=re.IGNORECASE)
m = regex.match('wesm@bright.net')
m.groups()
regex.findall(text)
print(regex.sub(r'Username: \1, Domain: \2, Suffix: \3', text))
```
