```python
import pandas as pd
import numpy as np
data = pd.Series(np.random.randn(9), index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'], [1,2,3,1,3,1,2,2,3]])
print(data)
print(data['b':'c'])
print(data[['b','d']])
#输出第二层index=2的值
print(data[:,2],'\n\n')
#unstack()函数重排数据为dataframe格式
data.unstack()
#stack()
data.unstack().stack()

frame = pd.DataFrame(np.arange(12).reshape((4, 3)),
                     index=[['a', 'a', 'b', 'b'], [1, 2, 1, 2]],
                     columns=[['Ohio', 'Ohio', 'Colorado'],
                              ['Green', 'Red', 'Green']])
print(frame)
frame.index.names = ['key1', 'key2']
frame.columns.names = ['state','color']
print(frame, '\n')
print(frame['Ohio'])
```

```python
print('########Reordering and Sorting Levels##########')
print(frame, '\n')
#swaplevel调换2个level的位置，但不影响数据值本身
print(frame.swaplevel('key1','key2'),'\n')
#sort_index可以根据单一的level来sort data
print(frame.sort_index(level=1),'\n')
print(frame.swaplevel(0,1), '\n')
print(frame.swaplevel(0,1).sort_index(level=0))
```

```python
print('########Summary Statistics by Level##########')
#针对不同的level进行数目统计, 输出table
print(frame.sum(level = 'key2'))
#对列特征color分类。列分类时候要设置axis = 1
print(frame.sum(level = 'color', axis = 1))
```

```python
print('########Indexing with a DataFrame’s columns##########')
frame = pd.DataFrame({'a': range(7), 'b': range(7, 0, -1),
                      'c': ['one', 'one', 'one', 'two', 'two',
                             'two', 'two'],
                      'd':[0,1,2,0,1,2,3]})
print(frame, '\n')
frame2 = frame.set_index(['c','d'])
print(frame.set_index(['c','d']), '\n')
print(frame.set_index(['c','d'], drop = False), '\n')
print(frame2.reset_index(), '\n')
```

#### 2 Combining and Merging Datasets

```python
print('########Database-Style DataFrame Joins##########')
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})
print(df1, '\n', df2, '\n')
#merge函数，没写按哪个列merge，建议加上on= , 而且default为两个data相交（intersect）。
print(pd.merge(df1, df2, on = 'key'))

df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                     'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                     'data2': range(3)})
#如果key名字不一样，可以用left_on= 和right_on=来指定。
print(pd.merge(df3, df4, left_on='lkey', right_on='rkey'), '\n')

#两个data的数据都要用 how = 'outer', 只要第一个data的用'left'，要两个data同时有的用'inner'
print(pd.merge(df1, df2, how = 'outer'),'\n')

#多对多merge
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
print(pd.merge(df1, df2, on='key', how='left'), '\n')

#根据多个keys来merge:
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                      'key2': ['one', 'two', 'one'],
                      'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                       'key2': ['one', 'one', 'one', 'two'],
                       'rval': [4, 5, 6, 7]})
print(pd.merge(left, right, on=['key1', 'key2'], how='outer'))
print(pd.merge(left, right, on = 'key1'),'\n')
print(pd.merge(left, right, on='key1', suffixes=('_left', '_right')))
```

```python
print('########Merging on Index##########')
import pandas as pd
import numpy as np
#可以用left_index=True和right_index=True 指示index作为merge key
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                       'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
print(pd.merge(left1, right1, left_on='key', right_index=True, how='outer'), '\n')

lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio','Nevada', 'Nevada'],
'key2': [2000, 2001, 2002, 2001, 2002],
'data': np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Nevada', 'Nevada', 'Ohio', 'Ohio',
                              'Ohio', 'Ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
print(pd.merge(lefth, righth, left_on=['key1','key2'], right_index=True),'\n')

#merge合并同时用两个data的index
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                      index=['a', 'c', 'e'],
                      columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                       index=['b', 'c', 'd', 'e'],
                       columns=['Missouri', 'Alabama'])
print(pd.merge(left2, right2, left_index=True, right_index=True),'\n')

#join()函数可以根据index合并2个或多个data，index一样活着类似
print(left2.join(right2, how='outer'), '\n')


another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                        index=['a', 'c', 'e', 'f'],
                        columns=['New York', 'Oregon'])
print(left2.join([right2,another]), '\n')
```

```python
print('########Concatenating Along an Axis##########')
arr = np.arange(12).reshape((3,4))
print(arr)
print(np.concatenate([arr,arr],axis=1),'\n')

#concat函数可以把多个series接起来
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
print(pd.concat([s1,s2,s3]), '\n')
#添加axis=1可以把series不合成一列，分列放
print(pd.concat([s1,s2,s3], axis=1),'\n')
#join_axes参数可以指定axes
s4 = pd.concat([s1, s3])
print(pd.concat([s1,s4],axis=1),'\n')
print(pd.concat([s1,s4],axis=1, join_axes=[['a','b','c','e']]),'\n')

#concat内的keys参数可以添加hierarchical index
result = pd.concat([s1,s2,s3], keys=['one','two','three'])
print(result, '\n\n')
print(result.unstack(), '\n')
#这里keys可以作为column header
print(pd.concat([s1,s2,s3], axis=1, keys=['one','two','three']),'\n')
#类似方法合并dataframe
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                    columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                    columns=['three', 'four'])
print(pd.concat([df1, df2], axis=1, keys=['level1','level2']), '\n')
print(pd.concat({'level1':df1, 'level2':df2},axis=1),'\n')
#用其他参数可以管理层次index
print(pd.concat([df1,df2], axis=1, keys=['level1','level2'], names=['upper','lower']), '\n')

#对于row index没有相关的data
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
print(pd.concat([df1, df2], ignore_index=True),'\n')
```

```python
print('########Combining Data with Overlap##########')
#np的where函数，执行array-oriented的if-else:
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
               index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
               index=['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan
print(np.where(pd.isnull(a), b, a), '\n')
#combine_first可以做到同样的结果,如果data1没有，则括号里的data2补上
print(b[:-2])
print(a[2:])
print(b[:-2].combine_first(a[2:]))

df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                     'b': [np.nan, 2., np.nan, 6.],
                     'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                     'b': [np.nan, 3., 4., 6., 8.]})
df1.combine_first(df2)
```

#### 3 Reshaping and Pivoting

```python
print('########Reshaping with Hierarchical Indexing##########')
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                     index=pd.Index(['Ohio', 'Colorado'], name='state'),
                     columns=pd.Index(['one', 'two', 'three'],
                     name='number'))
print(data,'\n')
result = data.stack()
print(data.stack(), '\n')
#unstack(0)可以从另一个level来做
print(result.unstack(0),'\n')
print(result.unstack('state'),'\n')

df = pd.DataFrame({'left': result, 'right': result + 5},
                   columns=pd.Index(['left', 'right'], name='side'))
df.unstack('state')
df.unstack('state').stack('side')
```

```python
print('########Pivoting “Long” to “Wide” Format##########')
data = pd.read_csv('../examples/macrodata.csv')
data.head()
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D','end')
print(data.head())
ldata = data.stack().reset_index().rename(columns={0:'value'})
print(ldata.head(),'\n')
pivoted = ldata.pivot('date', 'item', 'value')
print(pivoted.head(),'\n')

ldata['value2'] = np.random.randn(len(ldata))
pivoted = ldata.pivot('date','item')
print(pivoted.head())
```

```python
print('########Pivoting “Wide” to “Long” Format##########')
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                    'A': [1, 2, 3],
                    'B': [4, 5, 6],
                    'C': [7, 8, 9]})
print(df, '\n')
melted = pd.melt(df, ['key'])
print(melted, '\n')
reshaped = melted.pivot('key','variable','value')
print(reshaped, '\n')
print(reshaped.reset_index(),'\n')
print(pd.melt(df, id_vars=['key'], value_vars=['A', 'B']))
pd.melt(df, value_vars=['A', 'B', 'C'])
pd.melt(df, value_vars=['key', 'A', 'B'])
```
